## 6. Insights
1. High churn is strongly associated with Month-to-Month contracts

Customers on month-to-month plans have the highest churn rate compared to One-Year and Two-Year contracts.
This suggests that contract flexibility increases the likelihood of leaving.

2. Customers with higher Monthly Charges churn more often

Users paying more than $80/month show significantly higher churn.
Higher prices may create dissatisfaction or financial pressure, leading to cancellation.

3. Low-tenure customers are more likely to churn

Customers with less than 10 months of tenure have a higher churn probability.
This indicates early-cycle churn where customers leave shortly after joining.

4. Electronic Check users show the highest churn rate

Payment methods like Electronic Check have noticeably higher churn,
possibly because these customers also tend to have Month-to-Month contracts and higher fees.

5. Senior Citizens exhibit higher churn

Senior customers show a higher churn proportion compared to younger customers.
This may reflect higher price sensitivity or lower engagement with digital services.

# Recommendations
1. Create retention offers for Month-to-Month customers

Provide discounts, loyalty points, or temporary price reductions to encourage upgrading to One-Year contracts.

2. Reduce price pressure for high-charge customers

Introduce:

tiered plans

temporary discounts

personalized bundles

This helps lower churn among customers paying $80+ per month.

3. Improve onboarding for new customers (first 3–6 months)

Since low-tenure customers churn more:

send onboarding emails

offer product demos

give early-cycle discounts

provide proactive customer support

4. Investigate issues with Electronic Check payment users

Recommend switching users to:

Bank transfer

Credit card

Auto-payment systems

These methods correlate with lower churn.

5. Create targeted retention programs for Senior Citizens

Examples:

dedicated support line

simplified billing options

discounted senior plans

6. Use predictive modeling to identify high-risk churn customers

Now that you have cleaned and encoded data,
you can use it to train a logistic regression or random forest model to predict churn and intervene earlier.
